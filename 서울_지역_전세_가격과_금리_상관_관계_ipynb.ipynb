```python
import numpy as np

# 예측 전세가격과 실제 전세가격 데이터가 필요합니다.
# 예시로 예측값과 실제값을 임의로 생성하여 사용합니다.
forecasted_prices = np.array([1000, 1100, 1050, 1200, 1150, 1250, 1300, 1350, 1400, 1450, 1500, 1550])
actual_prices = np.array([980, 1050, 1000, 1150, 1100, 1200, 1250, 1300, 1350, 1400, 1450, 1500])

# 평균 절대 오차(MAE) 계산
mae = np.mean(np.abs(forecasted_prices - actual_prices))
print("평균 절대 오차(MAE):", mae)

# 평균 제곱 오차(MSE) 계산
mse = np.mean((forecasted_prices - actual_prices)**2)
print("평균 제곱 오차(MSE):", mse)

# 또는 각각의 예측값과 실제값을 비교하여 차이를 출력할 수도 있습니다.
for i in range(len(actual_prices)):
    print(f"실제 전세가격: {actual_prices[i]}, 예측 전세가격: {forecasted_prices[i]}, 오차: {forecasted_prices[i] - actual_prices[i]}")
```
